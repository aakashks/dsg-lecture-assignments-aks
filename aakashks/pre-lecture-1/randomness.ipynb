# Random walk
Xi+1 = Xi + epsilon i

```python
def walk(p, q, n):
    epsi = np.random.choice([p,q], size=n)
    x_arr=[epsi[:i].sum() for i in range(n)]
    return x_arr
```

```python
plt.plot(walk(-1, 1, 10000))
plt.title('Random Walk')
```

```python
plt.plot(walk(-2,10,10000))
plt.title('Irregular Walk')
```

```python
def normal(p, q, n):
    matrix = []
    for i in range(n):
        matrix.append(walk(p,q,n))
        
    dic = {}
    for i in range(-n//4, n//4+1):
        dic[i] = 0
    
    for i in range(n):
        for j in range(n):
            dic[matrix[i][j]] += 1
            
    return dic
            
```

```python
dic = normal(-1,1,1000)
```

```python
plt.plot(dic.keys(), dic.values())
plt.title('Central Limit Theory')
```

```python
for i in range(100):
    plt.plot(walk(-1,1,10000))

plt.title('Multiple Walks')
plt.xlabel('no of steps')
plt.ylabel('X')
plt.show()
```

Making random walk such that Xi+1 - Xi = ei
and ei ~ exp(1/lambda)

so now X will represent a stochastic process (which is poisson process)

```python
def poisson_process(scale, n):
    ei = np.random.exponential(scale, size=n)
    x_arr = [ei[:i].sum() for i in range(n)]
    return x_arr
```

```python
plt.plot(poisson_process(4, 10000))
plt.title('Poisson process stimulation')
plt.xlabel('no of steps')
plt.ylabel('X')
```

```python
def poisson_process_1(scale, n):
    ei = np.random.exponential(scale, size=n)
    x_arr = np.array([ei[:i].sum() for i in range(n)])
    Nx = [x_arr[:i].sum() for i in range(n)]
    return Nx
```

```python
plt.plot(poisson_process_1(3, 1000))
```

2D Random walk

```python
def walk_2d(n):
    x_arr = [0]
    y_arr = [0]
    for i in range(n):
        theta = np.random.uniform(0, 2 * np.pi)
        dx = np.cos(theta)
        dy = np.sin(theta)
        
        x_arr.append(x_arr[-1] + dx)
        y_arr.append(y_arr[-1] + dy)

    plt.plot(x_arr, y_arr)
    plt.title('2D walk')
    plt.xlabel('x')
    plt.ylabel('y')

    # return x_arr, y_arr
   
```

```python
walk_2d(2000)
```

```python
def walk_2d_1(n):
    theta_arr = np.random.random(n) * 2* np.pi
    arr = np.exp(theta_arr * 1j).cumsum()
    plt.plot(arr.real, arr.imag)
```
