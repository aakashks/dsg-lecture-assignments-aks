```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
def walk(p, q, n):
    epsi = np.random.choice([p,q], size=n)
    x_arr=[epsi[:i].sum() for i in range(n)]
    return x_arr
```

```python
plt.plot(walk(-2, 10, 10000))
```

```python
def normal(p, q, n):
    matrix = []
    for i in range(n):
        matrix.append(walk(p,q,n))
        
    dic = {}
    for i in range(-n//4, n//4+1):
        dic[i] = 0
    
    for i in range(n):
        for j in range(n):
            dic[matrix[i][j]] += 1
            
    return dic
            
```

```python
dic = normal(-1,1,1000)
```

```python
plt.plot(dic.keys(), dic.values())
```
